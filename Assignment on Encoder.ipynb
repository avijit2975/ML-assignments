# Label Encoder

```python
df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for column in df2.columns:
    if is_numeric_dtype(df2[column]):
        continue
    else:
        df2[column] = le.fit_transform(df2[column])
```

```python
df2.head()
```

# One Hot Encoder

```python
df3.head()
```

```python
col = ['sex','smoker','region']

for column in col:
    one = pd.get_dummies(df3[column],drop_first=True,prefix='dummy')
    df3 = pd.concat([df3,one],axis=1).drop(column,axis=1)
    
```

```python
df3.head()
```

# Ordinal Encoder

```python
df4.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal = OrdinalEncoder()
```

```python
cols = ['sex','smoker','region']

for col in cols:
    unique = df4[col].unique()
    df4[col] = OrdinalEncoder(categories=[unique]).fit_transform(df4[[col]])
```
